```python
import pandas as pd

class HondaServiceStation:
    print("\t\t\t\t\t\t\tHONDA SERVICE STATION")
    print("\t\t\t\t\t\t\t ESTABLISHED 2006")
    print("\t\t\t\t\t\t\t The Power of Dreams")

    def __init__(self):
        # Displaying available vehicle types
        x = ["bike", "car", "truck"]
        for k in x:
            print("\n\n", k)

        
        Vtype = input("Enter the vehicle type (car/bike/truck): ").lower()  # Input to determine vehicle type
        print(Vtype)     
        
        
        if Vtype == "car":
            self.car()
        elif Vtype == "bike":
            self.bike()
        elif Vtype == "truck":
            self.truck()
        else:
            print("Invalid vehicle type! Please enter car, bike, or truck.")

    def car(self):
        # Car service details
        x = ["Oil change", "Tyre change", "Brake Service", "Full service","Battery Replacement", "Air Filter replacement", " Spark Plug Replacement", "Coolant Filter service"]
        y = [100, 4000, 350, 12500,4000,3000,850,2500]
        car = {"Service type": x, "Service Charge": y}
        
        df = pd.DataFrame(car)
        print(df)
        
        d = []  # List to store the selected services' costs
        
        while True:
            try:
                pn = int(input("Enter the position number of the service required (0-3): "))
                
                if pn < 0 or pn >= len(car["Service Charge"]):
                    print("Invalid position number! Please enter a valid number.")
                    continue
                
                z = car["Service Charge"][pn]
                d.append(z)
                
                
                serv = input("Do you want any other type of service (yes/no): ").lower()
                
                if serv == "no":
                    break
            except ValueError:
                print("Invalid input! Please enter a number.")
        
        
        t = sum(d)
        m = t * 0.05  # 5% service tax
        k = t + m     # total cost + tax
        
        print("Service tax = 5%")
        print("Total Amount:", t, "Rupees")
        from colorama import Fore,Back,Style
        print(Fore.BLACK, Back.YELLOW, Style.BRIGHT, "Total Amount Including Service Tax: ",k,"Rupees")
        print(Fore.BLACK, Back.BLUE, Style.BRIGHT,"Thank you! Please visit again.")

    def bike(self):
        # Bike service details
        x = ["Oil change", "Tyre positioning and balancing", "Clutch Fluid", "Full service", "Brake Fluid", "Engine Repair", "Light repair", "Headstock bearing"]
        y = [300, 400, 350, 2250,200,1250,650,380]
        bike = {"Service type": x, "Service Charge": y}
        
        df = pd.DataFrame(bike)
        print(df)
        
        d = []  
        
        while True:
            try:
                pn = int(input("Enter the position number of the service required (0-3): "))
                
                if pn < 0 or pn >= len(bike["Service Charge"]):
                    print("Invalid position number! Please enter a valid number.")
                    continue
                
                z = bike["Service Charge"][pn]
                d.append(z)
                
                serv = input("Do you want any other type of service (yes/no): ").lower()
                
                if serv == "no":
                    break
            except ValueError:
                print("Invalid input! Please enter a number.")
        
        
        t = sum(d)
        m = t * 0.05  # 5% service tax
        k = t + m
        
        print("Service tax = 5%")
        print("Total Amount:", t, "Rupees")
        from colorama import Fore,Back,Style
        print(Fore.BLACK, Back.YELLOW, Style.BRIGHT, "Total Amount Including Service Tax: ",k,"Rupees")
        print(Fore.BLACK, Back.BLUE, Style.BRIGHT,"Thank you! Please visit again.")

    def truck(self):
        # Truck service details
        x = ["Engine Oil change", "Tyre Replacement", "Suspension Service", "Complete service"]
        y = [5000, 15000, 8000, 50000]
        truck = {"Service type": x, "Service Charge": y}
        
        df = pd.DataFrame(truck)
        print(df)
        
        d = []  
        
        while True:
            try:
                pn = int(input("Enter the position number of the service required (0-3): "))
                
                if pn < 0 or pn >= len(truck["Service Charge"]):
                    print("Invalid position number! Please enter a valid number.")
                    continue
                
                z = truck["Service Charge"][pn]
                d.append(z)
                
                serv = input("Do you want any other type of service (yes/no): ").lower()
                
                if serv == "no":
                    break
            except ValueError:
                print("Invalid input! Please enter a number.")
        
        
        t = sum(d)
        m = t * 0.05  # 5% service tax
        k = t + m
        
        print("Service tax = 5%")
        print("Total Amount:", t, "Rupees")
        from colorama import Fore,Back,Style
        print(Fore.BLACK, Back.YELLOW, Style.BRIGHT, "Total Amount Including Service Tax: ",k,"Rupees")
        print(Fore.BLACK, Back.BLUE, Style.BRIGHT,"Thank you! Please visit again.")



obj = HondaServiceStation()
```
